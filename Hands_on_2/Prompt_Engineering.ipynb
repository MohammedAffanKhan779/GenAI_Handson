```python
# Setup
from dotenv import load_dotenv
load_dotenv()

import getpass
import os

from langchain_google_genai import ChatGoogleGenerativeAI

if "GOOGLE_API_KEY" not in os.environ:
    os.environ["GOOGLE_API_KEY"] = getpass.getpass("Enter your Google API Key: ")
```

```python
# Low-ish temperature is good for code generation consistency
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.0,
)

# Structured prompt (CO-STAR style)
structured_prompt = """
# ROLE
You are a Senior Python Engineer writing production-grade Python code.

# TASK
Implement a Python function named `reverse_string_recursive` that:
- Accepts a single parameter: a string.
- Returns the reversed version of that string.
- Uses recursion only.

# STRICT REQUIREMENTS
1. You MUST use recursion.
2. You MUST NOT use:
   - String slicing with negative steps (e.g., s[::-1])
   - The reversed() function
   - Any built-in function that directly reverses sequences
3. The function must correctly handle:
   - Empty strings
   - Single-character strings
4. Follow PEP 8 naming and formatting conventions.
5. Code must be clean, readable, and production-ready.

# DOCUMENTATION REQUIREMENTS
Include a comprehensive docstring that contains:
- A clear description of the function’s purpose
- Parameters section with types
- Returns section with type
- One usage example
- Brief explanation of how recursion works in this function

# OUTPUT RULES (VERY IMPORTANT)
- Output ONLY valid Python code.
- Do NOT include markdown.
- Do NOT include backticks.
- Do NOT include explanations outside the function.
- Do NOT include example execution outside the docstring.

# TARGET AUDIENCE
Junior Python developers learning recursion and clean documentation practices.
"""
# Invoke the model with the structured prompt
response = llm.invoke(structured_prompt)

print(response.content)
```
